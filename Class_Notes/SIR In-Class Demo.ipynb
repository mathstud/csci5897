# In-Class Demo for SIR Model & Euler

We're going to implement the following simple recursion:
$$ t_{n+1} = t_n + \Delta t$$
$$ y_{n+1} = y_n + f(y_n, t_n) \cdot \Delta t$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SIR(S0,I0,R0, beta, gamma, t_max, stepsize):
    T = np.arange(0, t_max+stepsize, stepsize)
    S = np.zeros(len(T))
    I = np.zeros(len(T))
    R = np.zeros(len(T))
    N = S0+I0+R0

    for idx,t in enumerate(T):
        if idx == 0:
            S[idx] = S0
            I[idx] = I0
            R[idx] = R0
        else:
            dS_dt = -beta * S[idx-1] * I[idx-1] / N
            dI_dt = beta * S[idx-1] * I[idx-1] / N - gamma * I[idx-1]
            dR_dt = gamma * I[idx-1]

            S[idx] = S[idx-1] + dS_dt * stepsize
            I[idx] = I[idx-1] + dI_dt * stepsize
            R[idx] = R[idx-1] + dR_dt * stepsize
    
    return S, I, R, T
```

```python
# SIR(S0,I0,R0, beta, gamma, t_max, stepsize):
S,I,R,T = SIR(999,1,0,
             1,0.5,50,0.05)

fig, ax = plt.subplots()
ax.plot(T,S, color = 'b', label='Susceptibles')
ax.plot(T,I, color = 'r', label='Infecteds')
ax.plot(T,R, color = 'k', label='Recovereds')
ax.set_xlabel('time')
ax.set_ylabel('people')
ax.legend()
ax.set_ylim(bottom=0)
ax.set_xlim(left=0)
```

```python
# SIR(S0,I0,R0, beta, gamma, t_max, stepsize):
S,I,R,T = SIR(999,1,0,
             1.5,0.5,50,0.05)  # beta = how fast ppl get infected,  gamma = how fast ppl recover

fig, ax = plt.subplots()
ax.plot(T,S, color = 'b', label='Susceptibles')
ax.plot(T,I, color = 'r', label='Infecteds')
ax.plot(T,R, color = 'k', label='Recovereds')
ax.set_xlabel('time')
ax.set_ylabel('people')
ax.legend()
ax.set_ylim(bottom=0)
ax.set_xlim(left=0)
```

More infecteds faster than the first plot. Disease lasts twice as long by slowing recovery rate.
